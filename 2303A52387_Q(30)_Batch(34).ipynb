## Prediction of Online Shoppers Purchase Intentions.

### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('/content/online_shoppers_intention.csv')
```

### 1.Name the product most liked and purchased

```python
most_liked_product = data.loc[data['SpecialDay'].idxmax(), 'ProductRelated']


most_purchased_product = data.loc[data['Month'].idxmax(), 'ExitRates']

print(f"Most liked product: {most_liked_product}")
print(f"Most purchased product: {most_purchased_product}")
```

### 2. Identify the special days when most products are purchased

```python
special_days_purchases = data.groupby('SpecialDay')['Revenue'].sum()
print(special_days_purchases)

most_purchases_day = special_days_purchases.idxmax()
print(f"Most product purchases occur on Special Day: {most_purchases_day}")
```

### 3. List the top 5 special days of purchase and type of visitors.

```python
top_5_special_days = special_days_purchases.sort_values(ascending=False).head(5)
print("Top 5 Special Days for Purchases:")
print(top_5_special_days)

visitor_type_purchases = data.groupby(['SpecialDay', 'VisitorType'])['Revenue'].sum().reset_index()
top_visitor_types = visitor_type_purchases.sort_values(by='Revenue', ascending=False).groupby('SpecialDay').head(1)
print("\nTop Visitor Type for Each of the Top 5 Special Days:")
top_visitor_types
```

### 4. Identify the months when most purchases happened

```python
monthly_purchases = data.groupby('Month')['Revenue'].sum()

most_purchases_month = monthly_purchases.idxmax()

print(f"Most purchases happened in the month of: {most_purchases_month}")

print("\nPurchases per month:",monthly_purchases)
```

### 5. What the max and min revenue generated by the online shoppers

```python
max_revenue = data['Revenue'].max()
min_revenue = data['Revenue'].min()

print(f"Maximum revenue generated: {max_revenue}")
print(f"Minimum revenue generated: {min_revenue}")
```
